**Group 3: Hourly Activities**
Let's create visualization to review changes during different time frames within the day. 
I will start by creating the time frames first: *Night*, *Morning*, *Afternoon* and *Evening*.

```r
# Create a new column for the time of day
hourly_steps <- hourly_steps |> 
  mutate(time_of_day = case_when(
    hour(activity_hour) >= 0 & hour(activity_hour) < 6 ~ "Night",
    hour(activity_hour) >= 6 & hour(activity_hour) < 12 ~ "Morning",
    hour(activity_hour) >= 12 & hour(activity_hour) < 18 ~ "Afternoon",
    hour(activity_hour) >= 18 & hour(activity_hour) < 24 ~ "Evening"
  ))
```

```r
# Create a plot to visualize activity within daily time frames
hourly_steps |> 
  group_by(time_of_day) |> 
  summarise(mean_steps = mean(step_total)) |> 
  ungroup() |> 
  mutate(time_of_day = fct_reorder(time_of_day, mean_steps, .desc = TRUE)) |> 
  ggplot(aes(x = time_of_day, y = mean_steps)) +
  geom_col() +
  labs(title = "Average Steps per Time of Day", 
       subtitle = "Time Frames",
       x = "Time of Day",
       y = "Average Steps")
```

We may encourage for more steps in the mornings and in the evenings.

I would also like to break down to hourly periods, to see more details: 

```r
# Create a new column for the hour of the day
hourly_steps <- hourly_steps |> 
  mutate(hour_of_day = hour(activity_hour))
```

```r
# Create a plot to visualize activity within hourly periods
hourly_steps |> 
  group_by(hour_of_day) |> 
  summarise(mean_steps = mean(step_total)) |> 
  ggplot(aes(x = hour_of_day, y = mean_steps)) +
  geom_col() +
  labs(title = "Average Steps per Hour of the Day", 
       subtitle = "1-Hour Periods",
       x = "Hour of the Day",
       y = "Average Steps")
```

There seems to be a dip in between 2 and 4 pm. Tracker device may send a reminder to get a walk after the lunch. People could also use some boost *(such as* **gamification** *to compete with the community averages)* in the mornings to be more active. [4](###Recommendation.4)
